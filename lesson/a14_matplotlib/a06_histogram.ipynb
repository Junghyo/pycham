```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pylab as plt
from matplotlib import font_manager, rc
font_name= font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

```python
"""
히스토그램

1. 막대 그래프의 한 종류로 값의 [빈도]를 분리해서 보여준다.
2. 빈도를 처리하는 value_counts()
3. 숫자 데이터 hist( bins=@@) 
    1) bins : 구간의 개수 설정..

"""
```

```python
# 데이터 불러오기
%cd C:\pycharm\lesson\a14_matplotlib
df = pd.read_csv("lovefruits.csv", encoding="ms949")
df
```

```python
# 각 과일별 빈도수 구하기
freq = df["선호과일"].value_counts()
freq
```

```python
plt.bar(range(0, len(freq), 1), freq)
plt.xticks(range(0, len(freq), 1), freq.index)
plt.title("과일 선호도")
plt.show()
```

```python
df = pd.read_csv("student.csv", encoding="ms949")
df
```

```python
data = pd.concat([df['국어'],df['영어'],df['수학']])
print(data)
```

```python
# bins : 영역 나누기
plt.hist(data, bins=9)
plt.xticks(range(0,101,10), range(0,101,10))
plt.title("점수 빈도")
plt.show()
```

```python
plt.hist((df["국어"], df["영어"], df["수학"]), bins=10,
         label=["국어", "영어", "수학"])
plt.legend()
plt.show()
```
