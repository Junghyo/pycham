```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pylab as plt
import scipy as sp
```

```python
# making df
# 컬럼명은 다르지만 index(행)은 동일한 부분이 있도록
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3']},
                   index=['K0', 'K1', 'K2', 'K3'])

df2 = pd.DataFrame({'C': ['C2', 'C3', 'C4', 'C5'],
                    'D': ['D2', 'D3', 'D4', 'D5']},
                   index=['K2', 'K3', 'K4', 'K5'])
print(df1)
print(df2)
```

```python
# index를 기준으로 Left Join 하기 (Left join on index)
pd.merge(df1, df2, left_index=True, right_index=True,
         how="left")
```

```python
df1.join(df2, how="left")
```

```python
# index를 기준으로 Right Join 하기 (Right join on index)
pd.merge(df1, df2, left_index=True, right_index=True, how="right")
```

```python
df1.join(df2, how="right")
```

```python
# index를 기준으로 inner join 하기 (inner join on index)
pd.merge(df1, df2, right_index=True, left_index=True)
```

```python
df1.join(df2, how="inner")
```

```python
# index를 기준으로 outer join 하기 (outer join on index)
pd.merge(df1, df2, right_index=True, left_index=True, how="outer")
```

```python
# making df
# df1 의 key열의 값들과 df2의 인덱스 값들이 공통부분이 있도록

df1 = pd.DataFrame({'KEY': ['K0', 'K1', 'K2', 'K3'],
                    'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3']})
df2= pd.DataFrame({'C': ['C2', 'C3', 'C4', 'C5'],
                   'D': ['D2', 'D3', 'D4', 'D5']},
                  index=['K2', 'K3', 'K4', 'K5'])

print(df1)
print(df2)

```

```python
# index와 Key를 혼합해서 DataFrame 합치기 (Joining key columns on an index)
# df1의 key열과 df2의 index 혼합 merge
pd.merge(df1, df2, left_on="KEY", right_index=True, how="outer",
         indicator="info")
```

```python
df1.join(df2, on="KEY", how="outer")
```
