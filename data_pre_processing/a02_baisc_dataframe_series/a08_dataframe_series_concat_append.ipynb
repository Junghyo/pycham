```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pylab as plt
import scipy as sp
```

```python
df1 = pd.DataFrame(
    {
        'A': ['A0', 'A1', 'A2'],
        'B': ['B0', 'B1', 'B2'],
        'C': ['C0', 'C1', 'C2'],
        'D': ['D0', 'D1', 'D2']
    },
    index=[0, 1, 2]
)
# series에 name 설정
s1 = pd.Series(["S1", "S2", "S3"], name="S")

print(df1)
print(s1)
```

```python
# DataFrame에 Series '좌+우'로 합치기 : pd.concat([df, Series], axis=1)
# 새로운 컬럼명에 series의 이름이 들어간다.
# 설정을 안할 경우 0 : default
pd.concat([df1, s1], axis=1)
```

```python
# 기존 열 이름 무시하고 정수 번호 자동 부여 : ignore_index=True
pd.concat([df1, s1], axis=1, ignore_index=True)
```

```python
# making Series
s2 = pd.Series([0, 1, 2])
s3 = pd.Series([3, 4, 5])
print(s1)
print(s2)
print(s3)
```

```python
# Series 끼리 '좌+우'로 합치기 
# pd.concat([Series1, Series2, ...], axis=1)
# axis=1로 합치면 dataframe으로 변환
df2 = pd.concat([s1, s2, s3], axis=1)
print(type(df2))
df2
```

```python
# Series 끼리 합칠 때 열 이름(column name) 지정하기
# keys = ['xx', 'xx', ...]
pd.concat([s1, s2, s3], keys=["C0", "C1", "C2"], axis=1)
```

```python
# DataFrame에 Series를 '위+아래'로 합치기 
# df.append(Series, ignore_index=True)
# ignore_index=True 또는 Series에 name 설정
print(df1)
s1 = pd.Series(["S1", "S2", "S3", "S4"], 
               index=["A", "B", "C", "E"])
print(s1)
df1.append(s1, ignore_index=True)
```
