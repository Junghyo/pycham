```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pylab as plt
import scipy as sp
```

```python
# making df
np.random.seed(0)
df = pd.DataFrame(np.random.randn(5, 4), 
                  columns=["c1", "c2", "c3", "c4"])
df
```

```python
# inserting missing value
# index(0, 1), col(c1)
df.iloc[0:2, 0] = np.nan
# index(2), col(c2)
df.iloc[2, 1] = np.nan
df
```

```python
# 결측값이 들어있는 행 전체 삭제하기(delete row with NaN) 
# df.dropna(axis=0)
df1 = df.dropna(axis=0)
df1
```

```python
# 결측값이 들어있는 열 전체 삭제하기 (delete column with NaN) 
# df.dropna(axis=1)

df2 = df.dropna(axis=1)
df2
```

```python
# df[ ].dropna() : 특정 행 또는 열을 대상으로 결측값이 들어있으면 제거
# c1열의 결측치 제거
print(df.iloc[:, 0].dropna())
# c2열의 결측치 제거
print(df.iloc[:, 1].dropna())
# 0 index의 결측치 제거
print(df.iloc[0, :].dropna())
```
