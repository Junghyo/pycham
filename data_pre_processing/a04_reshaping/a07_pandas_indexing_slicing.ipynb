```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pylab as plt
import scipy as sp
import scipy.stats as stats
```

```python
# making series
s = pd.Series(np.arange(5), dtype="f")
s
```

```python
# indexing, slicing

# 1번째 값
print(s[0])
print(s.iloc[0])

# 1~3번째 index 값
print(s[1:4])
print(s.iloc[1:4])

# 전체 평균 이상인 값들만 return
print(s[s >= s.mean()])

# 4, 2, 0 index 선택
print(s[[4, 2, 0]])
print(s.iloc[[4, 2, 0],])
```

```python
# making series with index name
s1 = pd.Series([x for x in range(5)], dtype="f",
              index=["a", "b", "c", "d", "e"])
s1
```

```python
# a, b, e index
print(s1.loc[["a", "b", "e"]])

# c~e index
print(s1.loc["c":"e"])
print(s1.iloc[2:5])
print(s1.loc["c":])

# get() method : choose a, b, c index
print(s1.get([0, 1, 2]))
print(s1.get(["a", "b", "c"]))

```

```python
# add value, change value

# add
s1.loc["f"] = 100
s1["g"] = 50
s1
```

```python
# change
s1.loc["c"] = 777
print(s1)
s1.loc["d":"f"] = 111
print(s1)
```

```python
# checking index label
print("a" in s1)
print("x" in s1)

# checking values
print(777 in s1.values)
print(-1241 in s1.values)
```

```python
# DataFrame 행과 열 생성, 선택, 삭제 
# (creation, selection, drop of row and column)

# making df
df = pd.DataFrame({'C1': [0., 1., 2., 3.],
                   'C2': [4., 5., 6., 7.],
                   'C3': [8., 9., 10., np.nan]},
                  index=['R1', 'R2', 'R3', 'R4'])
print(df.index)
print(df.columns)
print(df.axes)
df

```

```python
# making new df (choose index or columns)

# r1, r3 index
df_r1r3 = pd.DataFrame(df, index=["R1", "R3"])
df_r1r3
```

```python
# c3, c1 columns
df_c3c1 = pd.DataFrame(df, columns=["C3", "C1"])
df_c3c1
```

```python
# r4, r2 index, c3, c1 columns
df_r4r2_c3c1 = pd.DataFrame(df, index=["R4", "R2"],
                            columns=["C3", "C1"])
df_r4r2_c3c1
```

```python
# selecting columns

# c3, c1
print(df.loc[:, ["C3", "C1"]])
print(df.iloc[:, [2, 0]])
print(df[["C3", "C1"]])
```

```python
# selecting rows(index)
# R1, R3
print(df.loc[["R1", "R3"], :])
```

```python
# selecting index and columns

# R1, R2 index, C1, C3 columns
print(df.loc["R1":"R2", ["C1", "C3"]])
```

```python
# using boolean

# c1 열이 1보다 작은 것들만 return
print(df[df.loc[:, "C1"] <= 1.0])

# C2열이 5초과인 경우만 RETURN
print(df[df.loc[:, "C2"] > 5.0])
```

```python
# 새로운 컬럼 만들기

# C5 = C1 + C2
df.loc[:, "C5"] = df.loc[:, "C1"] + df.loc[:, "C2"]
df
```

```python
# assign() method : C5 = C1 * C2
df = df.assign(C5 = df.loc[:, "C1"] * df.loc[:, "C2"])
df
```

```python
# assign() method using lambda
df = df.assign(C6 = lambda df: df.C1 * df.C2)
df
```

```python
# drop(), del df[] : 컬럼 삭제
print(df)

# drop(axis=1) : c5, c6 열 삭제
df1 = df.drop(labels=["C5", "C6"], axis=1)
df1
```

```python
# drop(axis=0) : 행 삭제
df1 = df.drop(labels=["R2", "R4"], axis=0)
df1
```

```python
# del df[]
df2 = df
print(df2)

# c2열 삭제
del df["C2"]
df
```
