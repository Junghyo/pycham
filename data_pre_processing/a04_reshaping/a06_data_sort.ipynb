```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pylab as plt
import scipy as sp
import scipy.stats as stats
```

```python
"""
- (1) DataFrame 정렬 : DataFrame.sort_values()
sort_values(
    by=None, : 정렬할 기준 변수
    axis=0, : 0 index, 1 columns
    ascending=True, : True 오름차순, False 내림차순
    inplace=False, : df 자체를 정렬해서 저장
    kind="quicksort", : 정렬 알고리즘
    na_position="last" L 결측값 위치(first, last)
)
 

- (2) Tuple 정렬 : sorted(tuple, key)

 

- (3) List 정렬 : list.sort(), sorted(list)

"""
```

```python
# making df
df = pd.DataFrame(
    {
        'sequence': [1, 3, 2],
        'name': ['park', 'lee', 'choi'],
        'age': [30, 20, 40]
    }
)

df
```

```python
# (1-1) 'sequence' 열(by='sequence')을 기준으로 
# index(axis=0) 오름차순 정렬하기
df.sort_values(by="sequence", axis=0)
```

```python
# (1-2) 내림차순(descending)으로 정렬하기 : ascending=False
df.sort_values(by="sequence", ascending=False)
```

```python
# (1-3) 열 이름을 (알파벳 순서로) 정렬하기 : sort_index(axis=1)
df.sort_index(axis=1, ascending=False)
```

```python
# index 정렬하기 : sort_index(axis=0)
df.sort_index(axis=0, ascending=False)
```

```python
# 결측값 위치 설정 : na_position= first or last
df.iloc[2, 2] = np.nan
df
```

```python
df.sort_values(by="sequence", na_position="first")
```

```python
df.sort_values(by="sequence", na_position="last")
```

```python
# (2) Tuple 정렬하기 : sorted(tuple, key) method
t = [(1, 'park', 30), (3, 'lee', 20), (2, 'choi', 40)]
print(t)

# sort by sequence
print(sorted(t, key=lambda x: x[0]))

# sort by name
print(sorted(t, key=lambda x: x[1]))

# sort by age, reverse=True : 역순
print(sorted(t, key=lambda x: x[2], reverse=True))
```

```python
# (3) List 정렬하기 : sorted(list), or list.sort()
np.random.seed(1)
l = list(np.random.randint(0, 100, 10))
print(l)
print(sorted(l))
# 역순
l.sort(reverse=True)
print(l)

```
