```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pylab as plt
import scipy as sp
import scipy.stats as stats
from sklearn.preprocessing import *
```

```python
# making df
df = pd.DataFrame(
    {
        'id' : ['C_001', 'C_001', 'C_002', 'C_002'],
        'prod' : ['P_001', 'P_002', 'P_001', 'P_002'],
        'cnt' : [1, 2, 3, 4],
        'price' : [100, 200, 300, 400]
    }, 
    columns=["id", "prod", "cnt", "price"]
)

df
```

```python
# (1) pd.melt(data, id_vars=['id1', 'id2', ...]) 를 사용한 데이터 재구조화
# id_vars에 들어간 컬럼명은 두고 
# 나머지 컬럼명은 variable 나머지컬렴의 데이터는 value
pd.melt(df, id_vars=["id", "prod"])
```

```python
# (2) pd.melt() 의 variable 이름, value 이름 부여하기 
# var_name, value_name
pd.melt(df, id_vars=["id", "prod"], var_name="info",
        value_name="info_value")
```

```python
df.pivot("id", "prod", "price")
```

```python
df_m = pd.melt(df, id_vars=["id", "prod"], var_name="info",
               value_name="info_value")
df_m
```

```python
df_p = pd.pivot_table(df_m, index=["id", "prod"], columns="info",
                      values="info_value")
df_p
```
