```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pylab as plt
import scipy as sp
import scipy.stats as stats
from sklearn.preprocessing import *
```

```python
# making df
data = pd.DataFrame(
    {
        'cust_id': ['c1', 'c1', 'c1', 'c2', 'c2', 'c2', 'c3', 'c3', 'c3'],
        'prod_cd': ['p1', 'p2', 'p3', 'p1', 'p2', 'p3', 'p1', 'p2', 'p3'],
        'grade' : ['A', 'A', 'A', 'A', 'A', 'A', 'B', 'B', 'B'],
        'pch_amt': [30, 10, 0, 40, 15, 30, 0, 0, 10]
    }
)
data
```

```python
# (1) 데이터 재구조화 : data.pivot(index, columns, values)
d_pivot = data.pivot(index="cust_id", columns="prod_cd", values="pch_amt")
d_pivot
```

```python
# index에 cust_id, grade를 설정할 경우
pd.pivot_table(data, index=["cust_id", "grade"], 
               columns="prod_cd",
               values="pch_amt")
```

```python
# columns가 2개인 경우 grade, prod_cd
pd.pivot_table(data, index="cust_id", columns=["grade", "prod_cd"],
               values="pch_amt")
```

```python
# with aggfunc
pd.pivot_table(data, index="grade", columns="prod_cd",
               values="pch_amt", aggfunc=np.sum)
```

```python
# groupby, unstack 사용
data.groupby(["prod_cd", "grade"]).sum().unstack("prod_cd")
```
