```python
# makeing df
df1 = pd.DataFrame(
    {
        'A': ['A0', 'A1', 'A2'],
        'B': ['B0', 'B1', 'B2'],
        'C': ['C0', 'C1', 'C2'],
        'D': ['D0', 'D1', 'D2']
    },
    index=[0, 1, 2]
)

df2 = pd.DataFrame(
    {
        'A': ['A3', 'A4', 'A5'],
        'B': ['B3', 'B4', 'B5'],
        'C': ['C3', 'C4', 'C5'],
        'E': ['E0', 'E1', 'E2']
    },
    index=[3, 4, 5]
)

print(df1)
print(df2)



```

```python
# 위 + 아래로 DataFrame 합치기(rbind) : axis = 0
pd.concat([df1, df2], axis=0)
```

```python
# 왼쪽 + 오른쪽으로 DataFrame 합치기(cbind) : axis = 1
pd.concat([df1, df2], axis=1)
```

```python
# 합집합(union)으로 DataFrame 합치기 : join = 'outer'
pd.concat([df1, df2], join="outer")
```

```python
# 교집합(intersection)으로 DataFrame 합치기 : join = 'inner'
pd.concat([df1, df2], join="inner")
```

```python
df3 = pd.DataFrame(
    {
        'A': ['A0', 'A1', 'A2'],
        'B': ['B0', 'B1', 'B2'],
        'C': ['C0', 'C1', 'C2'],
        'E': ['E0', 'E1', 'E2']
    },
    index=[0, 1, 3]
)

df3
```

```python
# axis=1일 경우 join outer
pd.concat([df1, df3], axis=1, join="outer")
```

```python
# axis=1일 경우 join inner
pd.concat([df1, df3], join="inner", axis=1)
```

```python
# axis=1일 경우 
# 특정 DataFrame의 index를 그대로 이용하고자 할 경우 : join_axes
pd.concat([df1, df3], axis=1, join_axes=[df1.index])
```

```python
df4 = pd.DataFrame(
    {
        'A': ['A0', 'A1', 'A2'], 
        'B': ['B0', 'B1', 'B2'],
        'C': ['C0', 'C1', 'C2'],
        'D': ['D0', 'D1', 'D2']
    },
    index=['r0', 'r1', 'r2']
)

df5 = pd.DataFrame(
    {
        'A': ['A3', 'A4', 'A5'],
        'B': ['B3', 'B4', 'B5'],
        'C': ['C3', 'C4', 'C5'],
        'D': ['D3', 'D4', 'D5']
    },
    index=['r3', 'r4', 'r5']
)
print(df4)
print(df5)


```

```python
# ignore_index=False : default
pd.concat([df4, df5], ignore_index=False)
```

```python
# ignore_index=True
pd.concat([df4, df5], ignore_index=True)
```

```python
# 계층적 index (hierarchical index) 만들기 : keys 
df45 = pd.concat([df4, df5], keys=["df4", "df5"])
df45
```

```python
# df45 indexing
# A열 보기
print(df45["A"])
print(df45.ix[:, 0])
print(df45.loc[:, "A"])
print(df45.iloc[:, 0])
```

```python
# df4행 전체 보기
print(df45.ix["df4"])
print(df45.loc["df4"])
print(df45.iloc[0:3])
```

```python
# df45의 r0, r1 보기
print(df45.ix["df4"][0:2])
print(df45.loc[("df4", ["r0", "r1"]), :])
print(df45.iloc[0:2, :])
```

```python
# index에 이름 부여하기 : names
df45.index.names = ["df_name", "row_number"]
df45.columns.name = "type"
df45
```

```python
# index 중복 여부 점검 : verify_integrity
pd.concat([df1, df3], verify_integrity=True)
# index에 중복이 있을 경우 에러가 뜸
```
