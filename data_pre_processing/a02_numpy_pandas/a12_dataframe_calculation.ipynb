```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pylab as plt
import scipy as sp
```

```python
# making df
df = pd.DataFrame(np.arange(10).reshape(5, 2),
                  index=["r1", "r2", "r3", "r4", "r5"],
                  columns=["c1", "c2"])
df
```

```python
# r2 index NaN처리
df.iloc[1] = None
# r3 index의 c2 열 NaN처리
df.iloc[2, 1] = None
# r5 index의 c1 열 NaN처리
df.iloc[4, 0] = None
df
```

```python
# mean(), std() 연산 시 : NaN은 분석 대상(population)에서 제외
print(df.mean()) # mean by columns
print(df.mean(axis=1)) # mean by rows
print(df.std())
# C1    3.333333 : (0+4+6)/3 = 3.333333
# C2    5.666667 : (1+7+9)/3 = 5.66667

```

```python
# DataFrame 칼럼 간 연산 시 : NaN이 하나라도 있으면 NaN 반환
df["c3"] = df["c1"] + df["c2"]
df
```
